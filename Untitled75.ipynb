```python
import pandas as pd

# Load the uploaded CSV file
file_path = '/content/test(in).csv'
df = pd.read_csv(file_path)

# Display the first few rows and the column names to understand the data structure
df.head(), df.columns.tolist()
```

```python
from textblob import TextBlob

# Function to label sentiment
def get_sentiment(text):
    if pd.isna(text):
        return 'Neutral'
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis on the 'body' column
df['sentiment'] = df['body'].apply(get_sentiment)

# Show sentiment distribution
df['sentiment'].value_counts()
```

```python
# Step 2: EDA and Sentiment Distribution Plot
sns.countplot(data=df, x='sentiment', order=['Positive', 'Neutral', 'Negative'])
plt.title('Sentiment Distribution')
plt.tight_layout()
plt.savefig('sentiment_distribution.png')
# Removed plt.clf() here to display the plot inline

# Step 3: Monthly Sentiment Scoring
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['month'] = df['date'].dt.to_period('M')
df['score'] = df['sentiment'].map({'Positive': 1, 'Neutral': 0, 'Negative': -1})
monthly_scores = df.groupby('month')['score'].mean().reset_index()

# Convert 'month' column to string for plotting
monthly_scores['month'] = monthly_scores['month'].astype(str)

sns.lineplot(data=monthly_scores, x='month', y='score', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sentiment Score')
plt.tight_layout()
plt.savefig('monthly_sentiment_score.png')
# Removed plt.clf() here to display the plot inline
```

```python
# Step 4: Employee Ranking by Positive Sentiments
employee_rank = df[df['sentiment'] == 'Positive']['from'].value_counts().reset_index()
employee_rank.columns = ['Employee', 'Positive_Emails']
employee_rank.to_csv('employee_ranking.csv', index=False)
```

```python
# Step 5: Flight Risk Identification
risk_df = df.groupby('from')['sentiment'].value_counts().unstack().fillna(0)
risk_df['risk_score'] = risk_df.get('Negative', 0) - risk_df.get('Positive', 0)
risk_df = risk_df.sort_values('risk_score', ascending=False)
risk_df.to_csv('flight_risk_employees.csv')
```

```python
# Step 6: Sentiment Trend Using Linear Regression
monthly_scores['month_num'] = LabelEncoder().fit_transform(monthly_scores['month'].astype(str))
model = LinearRegression()
model.fit(monthly_scores[['month_num']], monthly_scores['score'])
monthly_scores['trend'] = model.predict(monthly_scores[['month_num']])

sns.lineplot(data=monthly_scores, x='month', y='score', label='Actual')
sns.lineplot(data=monthly_scores, x='month', y='trend', label='Trend', linestyle='--')
plt.xticks(rotation=45)
plt.title('Sentiment Trend with Linear Regression')
plt.tight_layout()
plt.savefig('sentiment_trend_regression.png')
# Removed plt.clf() here to display the plot inline
```
